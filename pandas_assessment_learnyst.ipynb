```python
#Load the dataset in Pandas.
import pandas as pd

file = pd.read_csv("C:\\Users\\XNovaQ.18\\Downloads\\students_assignment_data(students_assignment_data).csv")
```

```python
file
```

```python
#Display first 15 rows, data types, and dataset shape.

print("First 15 \n")
print(file.head(15))

print("Data type \n")
print(file.dtypes)

print("Dataset shape \n")
print(file.shape)
```

```python
#Identify unique cities and count of students per city.

print("Unique cities \n")
print(file['City'].unique)

print("Count city \n")
print(file['City'].value_counts)
```

```python
#Filter and display students with Total_Marks greater than or equal to 400.

high_scorers = file[file['Total_Marks'] >= 400]

print(high_scorers)

print(high_scorers[['StudentID', 'Name', 'City', 'Total_Marks']])
```

```python
#Extract details of all female students and their scores.

gender = file[file['Gender'] == 'F']

print(gender[['StudentID', 'Name', 'City', 'Total_Marks']])
```

```python
#Find the average Total_Marks grouped by Gender.

avg_mrk_gender = file.groupby('Gender')['Total_Marks'].mean()

print(avg_mrk_gender)
```

```python
#Group students by City and compute the average Total_Marks and the total count per city

city_status = file.groupby('City').agg(Average_Total_Marks=('Total_Marks', 'mean'),Student_Count=('City', 'count'))

print(city_status)
```

```python
#Sort all students by Total_Marks in descending order.

sorted_students = file.sort_values(by='Total_Marks', ascending=False)

print(sorted_students)
```

```python
#Reset the index of the sorted DataFrame.

sorted_students = file.sort_values(by='Total_Marks', ascending=False)

sorted_students = sorted_students.reset_index(drop=True)

print(sorted_students)
```

```python
#Detect missing Total_Marks values (simulate by setting ~10% missing).

missing_marks = file[file['Total_Marks'].isna()]

print("Rows with Missing Total_Marks:\n")

print(missing_marks)
```

```python
#Impute missing Total_Marks with the mean of the column.

mean_marks = file['Total_Marks'].mean()

file['Total_Marks'] = file['Total_Marks'].fillna(mean_marks)

print("Missing values after imputation:", file['Total_Marks'].isna().sum())
```

```python
#Create a new column Result where students scoring >= 200 are marked as 'Pass', otherwise 'Fail'

file['Result'] = 'Fail'
file.loc[file['Total_Marks'] >= 200, 'Result'] = 'Pass'

print(file[['Total_Marks', 'Result']])
```

```python
#Count how many students failed and passed overall.

print(file['Result'].value_counts())
```

```python
file.to_csv("students_cleaned.csv")
```
